```python
# split the data infor quantile method
quantiles = RFMScore.quantile(q=[0.25,0.50,0.75])
quantiles = quantiles.to_dict()
quantiles
```

```python
def RecencyScore(x,p,d):
    if x<=d[p][0.25]:
        return 1
    elif x <=d[p][0.50]:
        return 2
    elif x <=d[p][0.75]:
        return 3
    else:
        return 4
```

```python
def FreqMonetryScore(x,p,d):
    if x<=d[p][0.25]:
        return 4
    elif x<=d[p][0.50]:
        return 3
    elif x<=d[p][0.75]:
        return 2
    else:
        return 1
```

```python
RFMScore.columns
```

```python
RFMScore['R'] = RFMScore['Recency'].apply(RecencyScore, args=('Recency',quantiles))
RFMScore['F'] = RFMScore['Frequency'].apply(FreqMonetryScore, args=('Frequency',quantiles))
RFMScore['M'] = RFMScore['Monetory'].apply(FreqMonetryScore, args=('Monetory',quantiles))
```

# Build ML Models

```python
def clean_transactions(df):
    df = df.copy()
    # fix column names
    if 'Date of purchase' in df.columns and 'Date' not in df.columns:
        df.rename(columns={'Date of purchase': 'Date'}, inplace=True)
    if 'InvoieNo' in df.columns and 'InvoiceNo' not in df.columns:
        df.rename(columns={'InvoieNo': 'InvoiceNo'}, inplace=True)

    # types
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df['Quantity'] = pd.to_numeric(df.get('Quantity', 1), errors='coerce').fillna(1)
    df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

    # line total
    df['LineTotal'] = (df['Price'] * df['Quantity']).fillna(0)

    # drop rows missing critical info
    df.dropna(subset=['CustomerID', 'InvoiceNo', 'Date', 'Price'], inplace=True)

    # remove future dates
    df = df[df['Date'] <= pd.Timestamp.now() + pd.Timedelta(days=1)]

    # remove duplicates
    df = df.drop_duplicates()

    # flag returns
    df['is_return'] = (df['LineTotal'] < 0).astype(int)

    return df

def churn_label_by_cutoff(df, cutoff_date, churn_days=90):
    cutoff_date = pd.to_datetime(cutoff_date)
    df_before = df[df['Date'] <= cutoff_date]
    df_after = df[(df['Date'] > cutoff_date) & (df['Date'] <= cutoff_date + pd.Timedelta(days=churn_days))]

    customers = pd.DataFrame(df['CustomerID'].unique(), columns=['CustomerID']).set_index('CustomerID')
    customers['purchases_after'] = df_after.groupby('CustomerID')['InvoiceNo'].nunique()
    customers['purchases_after'] = customers['purchases_after'].fillna(0).astype(int)
    customers['churn'] = (customers['purchases_after'] == 0).astype(int)  # 1 = churned

    return customers, df_before

def build_features_for_cutoff(df_before, cutoff_date, windows=[30,90,180]):
    cutoff_date = pd.to_datetime(cutoff_date)

    cust_agg = df_before.groupby('CustomerID').agg(
        last_purchase=('Date', 'max'),
        first_purchase=('Date', 'min'),
        freq_total=('InvoiceNo', 'nunique'),
        monetary_total=('LineTotal', 'sum'),
        avg_order_value=('LineTotal', 'mean'),
        return_count=('is_return', 'sum'),
    )

    cust_agg['recency_at_cutoff'] = (cutoff_date - cust_agg['last_purchase']).dt.days
    cust_agg['tenure_days'] = (cust_agg['last_purchase'] - cust_agg['first_purchase']).dt.days.fillna(0)

    for w in windows:
        start = cutoff_date - pd.Timedelta(days=w)
        sub = df_before[df_before['Date'] > start]
        g = sub.groupby('CustomerID').agg(
            freq_w=('InvoiceNo','nunique'),
            monetary_w=('LineTotal','sum'),
            avg_order_w=('LineTotal','mean'),
            uniq_items_w=('Item Code' if 'Item Code' in sub.columns else 'InvoiceNo','nunique')
        )
        g.columns = [f'{c}_{w}d' for c in ['freq','monetary','avg_order','uniq_items']]
        cust_agg = cust_agg.join(g, how='left')

    cust_agg.fillna(0, inplace=True)

    cust_agg['return_rate'] = cust_agg['return_count'] / cust_agg['freq_total'].replace(0, np.nan)
    cust_agg['return_rate'] = cust_agg['return_rate'].fillna(0)

    for col in ['recency_at_cutoff', 'tenure_days']:
        if pd.api.types.is_timedelta64_dtype(cust_agg[col]):
            cust_agg[col] = cust_agg[col].dt.days
        elif pd.api.types.is_datetime64_any_dtype(cust_agg[col]):
             # Calculate days from a reference point if it's a datetime column
             # Using the minimum date in the dataset as a reference
             min_date = df_before['Date'].min()
             cust_agg[col] = (cust_agg[col] - min_date).dt.days

    datetime_cols = cust_agg.select_dtypes(include=['datetime64[ns]', 'timedelta64[ns]']).columns
    if len(datetime_cols) > 0:
        print(f"Dropping non-numerical datetime/timedelta columns: {list(datetime_cols)}")
        cust_agg = cust_agg.drop(columns=datetime_cols)


    return cust_agg

def rolling_origin_evaluation(df, cutoff_dates, churn_days=90, windows=[30,90,180]):
    metrics = []
    models = []

    for i, cutoff in enumerate(cutoff_dates):
        customers_label, df_before = churn_label_by_cutoff(df, cutoff, churn_days=churn_days)
        X = build_features_for_cutoff(df_before, cutoff, windows=windows)

        data = X.join(customers_label[['churn']], how='inner')

        non_numerical_cols = data.select_dtypes(exclude=np.number).columns.tolist()
        if non_numerical_cols:
            print(f"Dropping non-numerical columns before training: {non_numerical_cols}")
            data = data.drop(columns=non_numerical_cols)


        if data['churn'].nunique() < 2 or len(data) < 100:
            print(f"Skipping cutoff {cutoff}: insufficient data or only one class present.")
            continue

        X_train, X_val, y_train, y_val = train_test_split(
            data.drop(columns='churn'),
            data['churn'],
            test_size=0.25,
            stratify=data['churn'],
            random_state=42
        )

        lgb_train = lgb.Dataset(X_train, label=y_train)
        lgb_val = lgb.Dataset(X_val, label=y_val)

        params = {
            'objective': 'binary',
            'metric': 'auc',
            'boosting': 'gbdt',
            'verbosity': -1,
            'seed': 42,
            'num_leaves': 31,
            'min_data_in_leaf': 50,
            'feature_fraction': 0.8,
            'bagging_fraction': 0.8,
            'bagging_freq': 5,
            'lambda_l1': 0.1,
            'lambda_l2': 0.1
        }

        callbacks = [
            lgb.early_stopping(stopping_rounds=50, verbose=False),
            lgb.log_evaluation(period=50)
        ]

        model = lgb.train(
            params,
            lgb_train,
            valid_sets=[lgb_train, lgb_val],
            num_boost_round=1000,
            callbacks=callbacks
        )

        prob = model.predict(X_val)
        auc = roc_auc_score(y_val, prob)

        metrics.append({'cutoff': cutoff, 'auc': auc, 'n': len(data)})
        models.append(model)

        print(f"Cutoff {cutoff} -> AUC {auc:.3f} (n={len(data)})")

    return metrics, models

df = pd.read_csv("E-com_Data.csv", low_memory=False)
df = clean_transactions(df)

last_date = df['Date'].max()
cutoffs = [last_date - pd.Timedelta(days=90*i) for i in range(4,0,-1)]

metrics, models = rolling_origin_evaluation(df, cutoffs, churn_days=90, windows=[30,90,180])
print("\nFinal metrics summary:")
print(metrics)
```

# Pipeline

```python
import pandas as pd
import numpy as np
import datetime as dt

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
import lightgbm as lgb

def rolling_origin_model_eval(df, cutoff_dates, churn_days=90, windows=[30,90,180]):
    results = []

    for cutoff in cutoff_dates:
        customers_label, df_before = churn_label_by_cutoff(df, cutoff, churn_days=churn_days)
        X = build_features_for_cutoff(df_before, cutoff, windows=windows)
        data = X.join(customers_label[['churn']], how='inner')

        if data['churn'].nunique() < 2 or len(data) < 200:
            print(f"Skipping cutoff {cutoff}: insufficient data")
            continue

        X_train, X_val, y_train, y_val = train_test_split(
            data.drop(columns='churn'),
            data['churn'],
            test_size=0.25,
            stratify=data['churn'],
            random_state=42
        )

        models = {
            "LogReg": LogisticRegression(max_iter=1000, class_weight="balanced"),
            "RandomForest": RandomForestClassifier(n_estimators=200, max_depth=10,
                                                   min_samples_leaf=20, random_state=42),
            "XGBoost": xgb.XGBClassifier(
                n_estimators=300, max_depth=6, learning_rate=0.05,
                subsample=0.8, colsample_bytree=0.8, random_state=42,
                eval_metric="auc", use_label_encoder=False
            ),
            "LightGBM": lgb.LGBMClassifier(
                n_estimators=300, num_leaves=31, learning_rate=0.05,
                subsample=0.8, colsample_bytree=0.8, random_state=42
            )
        }

        for name, model in models.items():
            model.fit(X_train, y_train)
            prob = model.predict_proba(X_val)[:,1]
            auc = roc_auc_score(y_val, prob)

            results.append({
                "cutoff": cutoff,
                "model": name,
                "auc": auc,
                "n": len(data)
            })

            print(f"Cutoff {cutoff.date()} | {name:12} | AUC={auc:.3f} | n={len(data)}")

    return pd.DataFrame(results)

df = pd.read_csv("E-com_Data.csv", low_memory=False)
df = clean_transactions(df)

last_date = df['Date'].max()
cutoffs = [last_date - pd.Timedelta(days=90*i) for i in range(4,0,-1)]

results_df = rolling_origin_model_eval(df, cutoffs, churn_days=90, windows=[30,90,180])

print("\nSummary of model performance:")
print(results_df.pivot(index="cutoff", columns="model", values="auc"))
```

```python
# app.py
import gradio as gr
import pandas as pd
import numpy as np
import datetime as dt
import lightgbm as lgb

def clean_transactions(df):
    df = df.copy()
    # fix column names
    if 'Date of purchase' in df.columns and 'Date' not in df.columns:
        df.rename(columns={'Date of purchase': 'Date'}, inplace=True)
    if 'InvoieNo' in df.columns and 'InvoiceNo' not in df.columns:
        df.rename(columns={'InvoieNo': 'InvoiceNo'}, inplace=True)

    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df['Quantity'] = pd.to_numeric(df.get('Quantity', 1), errors='coerce').fillna(1)
    df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

    df['LineTotal'] = (df['Price'] * df['Quantity']).fillna(0)

    df.dropna(subset=['CustomerID', 'InvoiceNo', 'Date', 'Price'], inplace=True)

    df = df.drop_duplicates()

    df['is_return'] = (df['LineTotal'] < 0).astype(int)

    return df


def build_features(df, cutoff_date=None, windows=[30,90,180]):
    if cutoff_date is None:
        cutoff_date = df['Date'].max()
    else:
        cutoff_date = pd.to_datetime(cutoff_date)

    cust_agg = df.groupby('CustomerID').agg(
        last_purchase=('Date', 'max'),
        first_purchase=('Date', 'min'),
        freq_total=('InvoiceNo', 'nunique'),
        monetary_total=('LineTotal', 'sum'),
        avg_order_value=('LineTotal', 'mean'),
        return_count=('is_return', 'sum'),
    )

    cust_agg['recency_at_cutoff'] = (cutoff_date - cust_agg['last_purchase']).dt.days
    cust_agg['tenure_days'] = (cust_agg['last_purchase'] - cust_agg['first_purchase']).dt.days.fillna(0)

    for w in windows:
        start = cutoff_date - pd.Timedelta(days=w)
        sub = df[df['Date'] > start]
        g = sub.groupby('CustomerID').agg(
            freq_w=('InvoiceNo','nunique'),
            monetary_w=('LineTotal','sum'),
            avg_order_w=('LineTotal','mean'),
            uniq_items_w=('Item Code' if 'Item Code' in sub.columns else 'InvoiceNo','nunique') # Handle missing 'Item Code'
        )
        g.columns = [f'{c}_{w}d' for c in ['freq','monetary','avg_order','uniq_items']]
        cust_agg = cust_agg.join(g, how='left')

    window_cols = [f'{c}_{w}d' for w in windows for c in ['freq','monetary','avg_order','uniq_items']]
    cust_agg[window_cols] = cust_agg[window_cols].fillna(0)


    # derived ratios
    cust_agg['return_rate'] = cust_agg['return_count'] / cust_agg['freq_total'].replace(0, np.nan)
    cust_agg['return_rate'] = cust_agg['return_rate'].fillna(0)

    for col in ['recency_at_cutoff', 'tenure_days']:
        if pd.api.types.is_timedelta64_dtype(cust_agg[col]):
            cust_agg[col] = cust_agg[col].dt.days
        elif pd.api.types.is_datetime64_any_dtype(cust_agg[col]):
             min_date = df['Date'].min()
             cust_agg[col] = (cust_agg[col] - min_date).dt.days

    cust_agg = cust_agg.drop(columns=['last_purchase','first_purchase'], errors='ignore')

    expected_features = [
        'recency_at_cutoff', 'tenure_days', 'freq_total', 'monetary_total',
        'avg_order_value', 'return_count', 'return_rate' # Base features
    ] + window_cols

    features_df = cust_agg.reindex(columns=expected_features, fill_value=0)


    for col in features_df.columns:
        features_df[col] = pd.to_numeric(features_df[col], errors='coerce')

    features_df = features_df.fillna(0)

    return features_df

try:
    model = lgb.Booster(model_file="churn_model.txt")
    print("LightGBM model loaded successfully.")
except Exception as e:
    model = None
    print(f"Error loading LightGBM model: {e}")
    print("Please ensure 'churn_model.txt' exists in the same directory as the app.py file.")

def predict_churn(file):
    """
    Processes uploaded CSV, builds features, and predicts churn probability
    for each customer using the loaded LightGBM model.
    """
    if model is None:
        return pd.DataFrame([{"Error": "Churn model not loaded. Cannot make predictions."}]), None

    try:
        df = pd.read_csv(file.name)
        df = clean_transactions(df)

        if df.empty:
             return pd.DataFrame([{"Message": "Cleaned data is empty. Cannot build features or predict."}]), None

        latest_date_in_file = df['Date'].max()
        features = build_features(df, cutoff_date=latest_date_in_file)


        if features.empty:
             return pd.DataFrame([{"Message": "Feature building resulted in an empty DataFrame."}]), None


        expected_feature_names = model.feature_name()

        features_aligned = features.reindex(columns=expected_feature_names, fill_value=0)

        features_aligned = features_aligned[expected_feature_names]


        preds_proba = model.predict(features_aligned)

        result = features_aligned.copy()
        result['CustomerID'] = features.index
        result['Churn_Probability'] = preds_proba

        result['Churn_Prediction'] = (result['Churn_Probability'] > 0.5).astype(int)

        result = result[['CustomerID', 'Churn_Probability', 'Churn_Prediction'] + expected_feature_names]

        out_path = "churn_predictions.csv"
        result.to_csv(out_path, index=False)

        return result, out_path # Return the full result DataFrame and the file path


    except Exception as e:
        print(f"An error occurred during prediction: {e}")
        return pd.DataFrame([{"Error": f"Prediction failed: {e}"}]), None

demo = gr.Interface(
    fn=predict_churn,
    inputs=gr.File(file_types=[".csv"], label="Upload Historical Transaction Data (CSV)"),
    outputs=[
        gr.Dataframe(label="Churn Predictions"),
        gr.File(label="Download Full Predictions CSV")
    ],
    title="Customer Churn Prediction",
    description="Upload your historical transaction data to get churn predictions for your customers."
)

if __name__ == "__main__":
    demo.launch(share=True)
```
